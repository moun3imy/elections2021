**1. Lire les données Excel et convertir les valeurs manquantes en 0 :**


```
import pandas as pd
import numpy as np
df = pd.read_excel("/content/parlement-elections-2016-1-1.xlsx").fillna(0)
df.head()
```

**Convertir les données float en int :**

```
for col in df.columns[-28:] : 
  df[col] = pd.to_numeric(df[col], errors='coerce',downcast='integer')
```

```
df['nSieges'] = pd.to_numeric(df['nSieges'], errors='coerce',downcast='integer')	
df['nInscrits'] = pd.to_numeric(df['nInscrits'], errors='coerce',downcast='integer')
df['pctNuls'] = pd.to_numeric(df['pctNuls'], errors='coerce',downcast='integer')
df['idSousPref'] = pd.to_numeric(df['idSousPref'], errors='coerce',downcast='integer')
```

```
df.head()
```

```
taux_participation_moyen  = df['txParticipation'].mean(skipna = True)
```

**`Taux de participation moyen est : `**





```
print(int(taux_participation_moyen*100))
```

**Nombre total de votes : **

```
df['Vote total']= df.iloc[:, -28:].sum(axis=1)
```
